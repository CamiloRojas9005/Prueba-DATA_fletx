```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ID,Nombre,Fecha,Costo.txt', index_col='ID')
df.head()
```

```python
df.dtypes  # tipos de datos iniciales
```

```python
df = df.drop_duplicates()   # eliminar los duplicados
df
```

```python
df = df.dropna()  # elimina valores nulos o nan
df
```

```python
df['Costo'] = pd.to_numeric(df['Costo'], errors='coerce')     # conversion a formato numerico
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')    # conversion a formato Time
df.dtypes
```

```python
df = df.dropna(subset=['Costo'])      # elimina los formatos que no correspondan a valores numericos o que sean nan
df = df.dropna(subset=['Fecha'])      # elimina los formatos que no corresponan al formato en tiempo o que sean nan
df
```

```python
# operaciones de suma y avg, tambien se hallan los maximos y minimos de la columna costo
suma = df['Costo'].sum()
avg = df['Costo'].mean()
minValue = df['Costo'].min()
maxValue = df['Costo'].max()
print(f'La Suma de los valores de la columna Costo es {suma} , el avg es {avg}, el valor minimo es {minValue} y el valor maximo es {maxValue}')
```

```python
#se hallan las fechas maximas y minimas de la columna Fecha
minDate = df['Fecha'].min()
maxDate = df['Fecha'].max()
print(f'La fecha minima es : {minDate}, y la fecha maxima es {maxDate}')
```

```python
#conteo de los productos
productos = df['Nombre'].value_counts()
productos
```

optimizar codigo para grandes cantidades de datos

```python
#lectura por bloques
# se realiza lectura por
chunkSize = 1000000
columnas = ['ID', 'Nombre', 'Fecha', 'Costo']

for chunk in pd.read_csv('/content/ID,Nombre,Fecha,Costo.CSV',chunksize=chunkSize, usecols=columnas):
  chunk.set_index('ID', inplace=True)
  chunk['Costo'] = pd.to_numeric(chunk['Costo'], errors='coerce')
  chunk['Fecha'] = pd.to_datetime(chunk['Fecha'], errors='coerce')
  chunk.dropna(subset=['Costo', 'Fecha'], inplace=True)

  suma2 = chunk['Costo'].sum()
  avg = chunk['Costo'].mean()
  contador = chunk.shape[0]

print(f'La suma de la columna suma es : {suma2}, el avg es : {avg} y el total de columnas procesadas es : {contador}')
```
